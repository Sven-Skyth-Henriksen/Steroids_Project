```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('anabolic.csv')
df
```

```python
# Data Converting
```

```python
df['Convertion_into_DHT'] = df['Convertion_into_DHT'].replace({'no':0, 'yes':1})
df['Derivative_of']= df['Derivative_of'].replace({'testosterone':0,'DHT':1})
df['Water_retention']= df['Water_retention'].replace({'no':0,'yes':1})
df['Aromatase']= df['Aromatase'].replace({'no':0,'yes':1})
df['Toxic']= df['Toxic'].replace({'weak':0,'strong':1})

df
```

```python
from jupyterthemes import jtplot
jtplot.style(sns.barplot(data=df, x="Toxic", y="Water_retention"))
```
